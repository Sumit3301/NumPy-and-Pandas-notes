# Data Visualization

## MatPlotlib
Matplotlib is an opensource library for Python programming language used for visualization for 2D plots of arrays which is built on NumPy. It allows visualization of large amount of data with easy and effecient usage. It allows python to be used as a scientific and mathematical usage, with NumPy embedded into it causes to to easily handle large amount of data with less time.

### Features of Matplotlib
- Support almost all Operating systems and supports different backend integration
- Time taken to visualize large data is less due to NumPy integration
- Allows us to visualize data with different visuals such as bar graph, pie chart, histogram

### Installation of matplotlib
#### Matplotlib can be insalled in MacOS, Windows and Linux using simple pip command
<blockquote> python -m pip install -U pip</blockquote>
<blockquote>python -m pip install -U matplotlib</blockquote>

### Importing Matplotlib

```python
from matplotlib import pyplot as plt
```

### pyplot in matplotlib
- pyplot is a collection of various functions in matplotlib similar to the MATLAB which enables usage of various visualization libraries of matlab.

### Plotting in matplotlib
- plot is an inbuilt function used to plot data
- Show function is used to display the plot, the plot is shown in a window

```python
#Plotting an array
import numpy as np
a=[1,2,3,4,5,6,7,8]
plt.plot(a)
plt.show()
```

- In matplotlib if we provide one list or an array to plot then it assumes that the values are for y values and automatically generates x values from 0 since the index starts from 0 of an array

### Plotting labelled data
- Labels help us to identify the data for different axis
- xlabel is used to define label for x axis
- ylabel is used to define label for the y axis
- name of the function is passed in both of the functions

```python
#Defining new plot
# countries=np.array([["United States","China","Japan","Germany","India"],[19.485,12.238,4.872,3.693,2.651]])
plt.plot(["United States","China","Japan","Germany","India"],[19.485,12.238,4.872,3.693,2.651])
# plt.plot(countries[0],countries[1])
plt.xlabel("Countries")
plt.ylabel("GDP")
plt.show()
#Defined GDP and labelled x and y axis accordingly
```

## Styles of plotting
- To change the limit of plots or to manipulate the plotting style there are various functions avalibale in pyplot
- One such function is axis(xmin,xmax,ymin,ymax)
- With this function we can change the limit of the graph shown to us

```python
plt.plot(([1,2,3,4],[6,3,8,6])) #Plotting with red objects as points
plt.axis([0,20,0,20]) # Defining limits of axis
plt.show()
```

- ### Changing the color of the plot

```python
np.array([["United States","China","Japan","Germany","India"],[19.485,12.238,4.872,3.693,2.651]])
plt.plot(["United States","China","Japan","Germany","India"],[19.485,12.238,4.872,3.693,2.651],color="green")
plt.xlabel("Countries")
plt.ylabel("GDP")
plt.show()
```

### Adding grids in graph
- Grids are structures made up of horizontal and vertial lines

```python
plt.plot(["United States","China","Japan","Germany","India"],[19.485,12.238,4.872,3.693,2.651],marker="o")
plt.xlabel("Countries")
plt.ylabel("GDP")
plt.grid()
plt.show()
```

### Subplot
Subplot allows us to plot more than 1 plot in a window, plotting more than 1 type of graph in a window allows us to analyze data from different perspetives at the same time. subplot() function is used to take advantage of it. Subplot is derived from MATLAB funtion
<blockquote>subplot(nrows, ncols, index, **kwargs)</blockquote>

```python
#Defining a numpy array
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
plt.subplot(1,2,1) #gprah on row 1
plt.plot(a,b)

c=np.array([12,11,7,30])
d=np.array([23,55,35,43])
plt.subplot(1,2,2)
plt.plot(c,d)
plt.show()
```

### Histograms
- It is a graph showing occurance withing each event. It gives us a broad idea of different events.
- hist() is used to plot a histogram

```python
histo=np.array([5,10,15,20,23,12,34,45,2,3,23,34,23,34,324,23,322])
fig, ax = plt.subplots(figsize =(10, 7))
lis=[]
for i in range(0,350,10):
    lis.append(i)
ax.hist(histo, lis)
```

### Pie charts
- Pie charts are the most common of analysing data
- pie() is used to draw a pie chart.
- pie chart draws a wedge for each part

```python
countries=["United States","China","Japan","Germany","India"]
gdp=np.array([19.485,12.238,4.872,3.693,2.651])

plt.pie(gdp,labels=countries)
plt.show()
```

 ### Adding legend
- To add more explanation we can also add legend by using legend function

```python
countries=["United States","China","Japan","Germany","India"]
gdp=np.array([19.485,12.238,4.872,3.693,2.651])

plt.pie(gdp,labels=countries)
plt.legend()
plt.show()
```

# Seaborn

Seaborn is an opensource python library based on matplotlib used for data visualization, it provides high level interface for stastical drawing.Seaborn makes it simple to visualize complex senarios and is an add on to matplotlib. It uses matplotlib to perform tasks. We can visualize large amount of data in a very short spam of time using seaborn

### Installing Seaborn
- Seaborn can be installed from anaconda prompt from Pypi command
<blockquote>pip install seaborn</blockquote>

### Importing Seaborn

```python
import seaborn as sns
```

### Loading a DataFrame in Seaborn
DataFrame is loaded in Seaborn library using load_dataset(), the default type of dataset is of CSV format

```python
import seaborn as sns
print(sns.get_dataset_names()) #print out the datasets already present with seaborn
df = sns.load_dataset("flights") #loading the titanic dataset and converting it into dataframe df
print(df) #printing out the dataframe
```

### Visualizing the dataset in seaborn

Seaborn is highly customizable for styling of graphs and unlike matplotlib not a lot of things needed to be done to get things going

```python
plt.scatter(df.year,df.passengers) #Displaying a scatter plot
sns.set() #Styling of graph
plt.show()
```

### Selecting different themes 
set_style() is used to declare different themes in seaborn,we can choose our own custom profile.Some of the themes are:
- Dark
- White
- Darkgrid

```python
#Visualizing by setting up theme
df1=sns.load_dataset("planets")
print(df1)
plt.scatter(df1.distance,df1.year)
sns.set_style("white") #c
plt.show()
```

### Plotting a graph from a dataframe
- relplot() function is used to plot the data, along with that it shows the relation between 2 variables with mapping of subsets

```python
sns.relplot(data=df1,x="distance",y="year")
```
